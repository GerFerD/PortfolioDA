## Cleaning the Data
At first glance, we can see that there are exactly 50,000 entries with twenty columns. The entries are comprised of two data types: "object" and "int64" (equivalent to Python's string/mixed and integer data types respectively). Five columns contain some null entries, which we might want to deal with. Many column labels are written in camelcase instead of the preferred snakecase. We will want to reformat the values under `prices` as integers while removing commas and dollar signs. Language might also be an issue, so it would probably be best to change German into English wherever possible.

We will begin by converting the columns to snakecase, and renaming some to improve descriptiveness.

```python
autos.columns
```

```python
autos.columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price_eur', 'ab_test',
                 'vehicle_type', 'registration_year', 'transmission', 'power_ps', 'model',
                 'odometer_km', 'registration_month', 'fuel_type', 'brand','unrepaired_damage',
                 'ad_created', 'n_of_pictures', 'postal_code', 'ad_last_seen']
```

```python
autos.head()
```

In addition to the conversion to snakecase, we chose to make several changes to the column names:
1. `price` to `price_eur`. The plan is to remove the dollar signs ("$") and commas from entries under this label in order to convert them into integers. After checking the website, *eBay Kleinanzeigen*, we confirmed that prices are in Euros.

2. `gearbox` to `transmission`. A gearbox is different from transmission in english, and the column entries were clearly transmission types.

3. `odometer` to `odometer_km`. Like the `price_eur` column, we will be converting these entries into integers.

4. `lastSeen` to `ad_last_seen`. The data entries are of advertisements of used cars, and "`lastSeen`" is not a clear enough description for the column.

```python
autos.describe(include='all')
```

The columns labeled `seller`, and `offer_type` are cadidates for complete removal. They both contain 2 unique values, one of which exists in 1/50000 entries.

Several columns require more investigation, especially ones like `power_ps` and `n_of_pictures` which should be simple numbers but are returning null values.

`price_eur`, `registration_year`, `odometer_km`, etc. will be converted to integers to allow for calculations.

### Cleaning and Exploring `price_eur` and `odometer_km`
In order to convert the column entries into integers, we must first remove dollar signs($) and commas.

```python
autos['price_eur'] = autos['price_eur'].str.replace('$','')
autos['price_eur'] = autos['price_eur'].str.replace(',','')
autos['price_eur'] = autos['price_eur'].astype(int)

autos['odometer_km'] = autos['odometer_km'].str.replace('km','')
autos['odometer_km'] = autos['odometer_km'].str.replace(',','')
autos['odometer_km'] = autos['odometer_km'].astype(int)
```

#### `price_eur'
We will now explore the same two columns, beginning with the price column, to look for any outliers in the data:

```python
autos['price_eur'].unique().shape
```

```python
autos['price_eur'].describe()
```

```python
price_counts = autos['price_eur'].value_counts()
price_counts.sort_index()
```

Right away, we can see that there are both high and low priced outliers. From looking at current used car listings on eBay, it is clear that some numbers (low and high) are used as pre-negotiation placeholders, and are not the actual price of the car. Several low price entries are also for junk cars to be broken down and used for parts.

In order to facilitate our analysis we will be removing all listings below 500 euros, as they are not likely to be selling functioning cars. Anything above 350,000 euros will also be removed because there is a significant jump in cost for the next lisiting at exactly 999,990 euros.

```python
autos = autos[autos['price_eur'].between(500,350000)]
```

```python
autos['price_eur'].unique().shape
```

```python
autos['price_eur'].describe()
```

```python
price_counts = autos['price_eur'].value_counts()
price_counts.sort_index()
```

With the new parameters set, our data now contains 45,097 entries, nearly 5,000 fewer entries than before.

At first glance the new dataframe might appear to still have many outliers. 75% of all entries are priced at 7,900 euros or less, while the most expensive listing is 350,000 euros.

Upon deeper investigation (below), however, we can see that although there are clearly high-priced entries and potential outliers, these data points are not erroneous and provide important information.

For example, the most expensive listing is for a Porsche 991 costing 350,000 euros. A quick internet search in March of 2021 (5 years after this entry was listed) yielded several cars of the same make and model being sold for around 200,000 euros. Considering the age of our data, as well as the introduction of newer Porsche models, this Porsche 991 entry is likely not an erroneous outlier we can simply remove.

```python
autos[autos['price_eur'] == 350000]
```

#### `odometer_km`
Below, we will similarly analyze the "odometer_km" column:

```python
autos['odometer_km'].unique().shape
```

```python
autos['odometer_km'].describe()
```

```python
autos['odometer_km'].value_counts().sort_index()
```

Immediately, we encounter three problems with the `odometer_km` column:

1. There are only thirteen unique entries. These correspond with eBay Kleinanzeigen's search engine filters. Instead of providing the number of kilometers on each car, it tells us what range the entry falls under. For example, there are 808 entries which have the "odometer_km" value of "40000". This means that 808 vehicles have anywhere between 30,000 and 40,000 kilometers on them. This might not be especially useful because the ranges are not of the same magnitude.

2. The maximum mileage for the column is 150,000 kilometers. This heavily skews the data because over 75% of all entries fall within this range (125,000 - 150,000 km). 

3. There are many cars on the website which have over 150,000 kilometers on them, but there is no category that includes them. Perhaps this means there are no entries in our data that have over 150,000 kilometers. It could instead mean that everything above 150,000 was lumped into the same category. There is no way to know. Although based on how heavily skewed the data is, we cannot rule out the latter.

We are now left with two options. We could either:

* Convert entries in `odometer_km` to strings and add some information (e.g. "0-5000" instead of "5000"). This would limit our ability to do mathematical computations using the data, and would blindly assume that there are no vehicles in the data that have more than 150,000 km on them, despite the website having several today.
* Or we could remove the entries above "125000" in order to eliminate assumptions. This, however, would remove 28,698 entries.

```python
autos = autos[autos['odometer_km'].between(0,125000)]
```

```python
autos['odometer_km'].unique().shape
```

```python
autos['odometer_km'].describe()
```

```python
autos['odometer_km'].value_counts().sort_index()
```

As shown above, we opted to remove the entries above "125000" from the `odometer_km` column, as it was the simpler option.

```python
autos.info()
```

Here we noticed the indeces are no longer matching the length of our dataset, so it was quickly adjusted below.

```python
autos = autos.set_axis(list(range(0,len(autos))))
```

There are several columns whose values represent dates:

Entries in `date_crawled`, `ad_created`, and `ad_last_seen` are all strings. These will need to be changed so we can analyze them. The `registration_month` and `registration_year` columns are already represented as integers.

First we will examine the format of the entries in the three columns:

```python
autos[['date_crawled','ad_created','ad_last_seen']].head()
```

```python
autos[['date_crawled','ad_created','ad_last_seen']].tail()
```

All entries appear to be written in the same format ("yyyy-mm-dd hh-mm-ss"). In order to look at the relative frequencies of the dates, we will only use at the first 10 characters of each entry (`.str[:10]`) and make sure to include non numerical entries (`dropna=False`). The relative frequencies will be shown as percentages instead of ratios.

```python
(autos['date_crawled'].str[:10].value_counts(normalize=True, dropna=False)*100).sort_index()
```
